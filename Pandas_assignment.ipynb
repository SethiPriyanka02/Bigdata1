## Q25. What is the difference between .loc and .iloc in Pandas?

The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc(). (loc: loc is used for label-based slicing)

The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().(iloc: iloc is used for position based slicing)

```python
# selecting rows from 1 to 4 and columns from 2 to 4
display(data.iloc[1: 5, 2: 5])
```

```python
display(data.loc[(data.City == 'Fort Lauderdale') & (data.Quantity==2)])
```
